# Storytelling Data Visualization on Exchange Rates

### Introducing the Dataset

In these project we will use [the dataset](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020), created by Daria Chemkaeva, which describe Euro daily exchange rate between 1999 and 2021. The euro (symbolized with €) is the official currency in most of the countries of the European Union.

If the exchange rate of the euro to the US dollar (symbolized with $) is 1.5, you get 1.5 $ if you pay 1.0 € (one euro has more value than one US dollar at this exchange rate). 

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
# enables Jupyter to display graphs
%matplotlib inline 

# set the style to "fivethirtyeight" for the matplotlib plots
style.use(style="fivethirtyeight")
```

```python
# importing data
exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
# display the first few rows of the exchange_rates DataFrame
display(exchange_rates.head(5))
```

```python
# display the last few rows of the exchange_rates DataFrame
display(exchange_rates.tail())
```

```python
# printing information about yhe exchange_rates DataFrame
display(exchange_rates.info())
```

As we can see in abowe table the dataset contain 41 columns and 5699 rows. In few colums (for example: `[Bulgarian lev ]` and `[Greek drachma ]`) we have null values (null values are present when `Non-Null Count` is different than numbers of rows). Also in dataset we have 3 colums with numeric data type and 38 columns with object data type.

###  Data Cleaning

We need to do data cleaning and prepare it for analysis.

In fist data cleaning step we will:
- rename specific columns: `Period\Unit:` and `Us dollar "
- removed extra whitespace and special characters from columns names
- add underscores ('_') to join names together
- converted columns names to title case

```python
# rename specific column
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar', 'Period\\Unit:': 'Time'}, inplace=True)

# rename columns by removing special characters using regex
exchange_rates.columns = exchange_rates.columns.str.replace(pat=r"[(^\[)(\]$)]", repl="", regex=True).str.strip()

# replace spaces with underscores and  converted columns names
exchange_rates.columns = exchange_rates.columns.str.replace(pat=" ", repl="_")
```

In second step we will:
- converted the 'Time' column to a datetime dtype
- sorted the exchange_rates Dataframe by the 'Time' column in ascending order
- reset the index in the exchange_rates Dataframe

```python
# convert 'Time' column to datetime
exchange_rates["Time"] = pd.to_datetime(exchange_rates["Time"])

# sort DataFrame by 'Time' column
exchange_rates.sort_values("Time", inplace=True)

# reset index after sorting
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
display(exchange_rates)
```

### Data processing and exploration

The following process will be followed:

- A subset of the original dataset will be extracted.
- All irrelevant/missing values will be removed.
- Column dtypes will be modified as required.

```python
# extracting the "Time" and "US_Dollar" columns from the exchange_rates DataFrame
euro_to_dollar = exchange_rates[["Time", "US_dollar"]].copy()
```

```python
# returning a Series containing counts of unique values in descending order
print(euro_to_dollar['US_dollar'].value_counts())
```

As we can see there are 62 `"-"` characters in our dataset, so we need to remove them.

```python
# removing from dataset rows with '-' characters
euro_to_dollar = euro_to_dollar[euro_to_dollar["US_dollar"] != '-']

# convert the "US_Dollar" column to numeric values
euro_to_dollar["US_dollar"] = euro_to_dollar["US_dollar"].astype(float)

# printing information about yhe exchange_rates DataFrame
euro_to_dollar.info()
```

### Rolling Mean

After cleaning the Data, we're generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.xlabel("Time")
plt.ylabel("US dollar exchange rate")
plt.title("The evolution of the euro-dollar exchange rate in time")
plt.show()
```

The wiggles are the visual representation of the daily variation in the exchange rate. The rate only shows clear upward or downward trends in the longer run (months or years).

We may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average). The rolling window can take various values.

Below, we see how line plot changes visually as we increase the rolling window. The window gets larger, and the line becomes smoother. 

```python
# create a plot consisting of plots showing data in rollin window
plt.figure(figsize=(9,6)) 

# creat a orginal value plot
plt.subplot(3,2,1) 
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

# creat plots in different rolling window
for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

```python
# calculate the rolling mean of the US Dollar exchange rate over a window of 30 days
# rounding the rolling mean to 2 decimal places for querying
euro_to_dollar["Rolling_mean"] = round(euro_to_dollar["US_dollar"].rolling(window=30).mean(), 2)
```

```python
# creating plot of euro to dollar DAtaFrame in 30 days rolling window
plt.plot(euro_to_dollar["Time"], euro_to_dollar["Rolling_mean"])
plt.show()
```

### Coming up with an Idea

We want to show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)).

### Sketching a Graph for the euro-dollar rate change under the last three US president

```python
# creating dataset containing data on the euro-dollar change under the last three presidents
three_presidents = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2001)
                                         & (euro_to_dollar["Time"].dt.year < 2021)]

# creat dataset for the euro-dollar change under the every presidents
bush = three_presidents.copy()[three_presidents["Time"].dt.year < 2009]
obama = three_presidents.copy()[(three_presidents["Time"].dt.year >= 2009) 
                                & (three_presidents["Time"].dt.year < 2017)]
trump = three_presidents.copy()[(three_presidents["Time"].dt.year >= 2017) 
                                & (three_presidents["Time"].dt.year < 2021)]
```

```python
euro_to_dollar_mean = round(three_presidents["US_dollar"].mean(), 2)
print("Euro to Dollar exchange rate mean value under three last US presidents:", str(euro_to_dollar_mean))
```

We used matplotlib's functional approach to build the graphs. We use this approach because it offers more flexibility in arranging the subplots:

- first we build three of the graphs on a 2-by-3 grid
- then we build summarize graph of a 2-by-1 grid
- we use `get_xticks` and `get_yticks` to find coordinates for text

```python
# adding the subplots
plt.figure(figsize=(12,7))

ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 1, 2)
axes = [ax1, ax2, ax3, ax4]

# changes apply to all the subplots
for ax in axes:
    ax.set_ylim(0.7, 1.7)
    ax.set_yticks([0.8, 1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['0.8', '1.0', '1.2','1.4', '1.6'],
                   alpha=0.4)
    ax.grid(alpha=0.5)
    
# creating subplot for Bush
ax1.plot(bush["Time"], bush["Rolling_mean"], c="#DA70D6")
ax1.set_xticklabels(["", "2001", "", "2003", "", "2005", "", "2007", "", "2009"], alpha=0.4)
ax1.text(731516.0, 1.85, "Bush", size=18, weight="bold", c="#DA70D6")
ax1.text(731116.0, 1.75, "(2001-2009)", size=15, weight="bold", c="#DA70D6", alpha=0.5)

# creating subplot for Obama
ax2.plot(obama["Time"], obama["Rolling_mean"], c="#66CDAA")
ax2.set_xticklabels(["", "2009", "", "2011", "", "2013", "", "2015", "", "2017"], alpha=0.4)
ax2.text(734288.0, 1.85, "Obama", size=18, weight="bold", c="#66CDAA")
ax2.text(734088.0, 1.75, "(2009-2017)", size=15, weight="bold", c="#66CDAA", alpha=0.5)

# creating subplot for Trump
ax3.plot(trump["Time"], trump["Rolling_mean"], c="#4682B4")
ax3.set_xticklabels(["2017", "", "2018", "", "2019", "", "2020", "", "2021"], alpha=0.4)
ax3.text(736855.0, 1.85, "Trump", size=18, weight="bold", c="#4682B4")
ax3.text(736655.0, 1.75, "(2017-2021)", size=15, weight="bold", c="#4682B4", alpha=0.5)

# creating subplot for three presidents
ax4.plot(bush["Time"], bush["Rolling_mean"], c="#DA70D6")
ax4.plot(obama["Time"], obama["Rolling_mean"], c="#66CDAA")
ax4.plot(trump["Time"], trump["Rolling_mean"], c="#4682B4")
ax4.set_xticks([])

# adding title and subtitle
ax1.text(730416.0, 2.20, '''EURO-USD rate average 1.22 under the last three US presidents''',
         fontsize=20, weight='bold')
ax1.text(730316.0, 2, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017) 
and Donald Trump (2017-2021)''', fontsize=16, alpha=0.5)

# adding a signature
ax4.text(729850.0, 0.65, 'Klaudia Syc' + ' '*110 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=14)

plt.show()
```

#### Idea

We will show how the euro-zloty rate changed during the 2007-2008 financial crisis. Also we will show the data for 2006 and 2009 for comparison.

```python
financial_crisis = euro_to_zloty.copy()[(euro_to_zloty["Time"].dt.year >= 2007) 
                                        & (euro_to_zloty["Time"].dt.year <= 2008)]
financial_crisis_6_9 = euro_to_zloty.copy()[(euro_to_zloty["Time"].dt.year >= 2006) 
                                        & (euro_to_zloty["Time"].dt.year <= 2009)]
```

```python
print("Max euro-zloty rate:", str(financial_crisis_6_9["Rolling_mean"].max()))
```

```python
print("Min euro-zloty rate:", str(financial_crisis_6_9["Rolling_mean"].min()))
```

```python
# adding a plot
fig, ax = plt.subplots(figsize=(8,3.5))
ax.plot(financial_crisis_6_9["Time"], financial_crisis_6_9["Rolling_mean"], linewidth=1, color="#32CD32")

# highlighting the 2007-2008 period
ax.plot(financial_crisis["Time"], financial_crisis["Rolling_mean"], linewidth=3, color="#e23d28")

# highlihting the hole of the crisis
ax.axvspan(xmin=733150.0, xmax=733350.0, ymin=0,
           alpha=0.3, color="grey")

# adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 732272.0
for year in ["2006", "2007", "2008", "2009", "2010"]:
    ax.text(x, 3.1, year, alpha=0.5, fontsize=11)
    x += 365

y = 3.22  
for rate in ["3.25", "3.50", "3.75", "4.00", "4.25", "4.50"]:
    ax.text(732172.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.25

# adding a title and a subtitle    
ax.text(732225.0, 4.88, "Euro-PLN rate hole at 3.25 during 2007-2008's financial crisis",
        weight='bold')
ax.text(732550.0, 4.75, 'Euro-PLN exchange rates between 2006 and 2010',
        size=12)

# adding a signature
ax.text(732172.0, 2.9, 'Klaudia Syc' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

# add some transparency to the grid
ax.grid(alpha=0.5)

plt.show()
```
