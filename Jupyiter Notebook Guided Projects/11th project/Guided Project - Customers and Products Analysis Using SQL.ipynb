### Question 1: Which Products Should We Order More of or Less of?

This question refers to inventory reports, including low stock(i.e. product in demand) and product performance. This will optimize the supply and the user experience by preventing the best-selling products from going out-of-stock.
- The low stock represents the quantity of the sum of each product ordered divided by the quantity of product in stock. We can consider the ten highest rates. These will be the top ten products that are almost out-of-stock or completely out-of-stock.
- The product performance represents the sum of sales per product.
- Priority products for restocking are those with high product performance that are on the brink of being out of stock.

```python
Image(url= "https://dq-content.s3.amazonaws.com/600/products-orders-tables.png")
```

Compute the low stock for each product

```python
%%sql
SELECT p.productCode, p.productName, od.quantityOrdered, p.quantityInStock,
       ROUND(SUM(od.quantityOrdered) * 1.0 / p.quantityInStock, 2) AS low_stock
  FROM products AS p
  JOIN orderdetails AS od
    ON p.productCode = od.productCode
 GROUP BY p.productCode
 ORDER BY low_stock DESC
 LIMIT 10;
```

Compute the product performance for each product

```python
%%sql
SELECT p.productCode, p.productName, od.quantityOrdered, od.priceEach,
       ROUND(SUM(od.quantityOrdered * od.priceEach), 2) AS product_performance
  FROM orderdetails AS od
  JOIN products AS p
    ON od.productCode = p.productCode
 GROUP BY p.productCode
 ORDER BY product_performance DESC
 LIMIT 10;
```

Display priority products for restocking 

```python
%%sql
WITH 
low_stock_table AS (
    SELECT p. productCode
      FROM (SELECT p.productCode, p.productName, od.quantityOrdered, p.quantityInStock,
                   ROUND(SUM(od.quantityOrdered) * 1.0 / p.quantityInStock, 2) AS low_stock
              FROM products AS p
              JOIN orderdetails AS od
                ON p.productCode = od.productCode
             GROUP BY p.productCode
             ORDER BY low_stock DESC) AS p
),
product_performane_table AS (
    SELECT p.productCode
      FROM (SELECT p.productCode, p.productName, od.quantityOrdered, od.priceEach,
                   ROUND(SUM(od.quantityOrdered * od.priceEach), 2) AS product_performance
              FROM orderdetails AS od
              JOIN products AS p
                ON od.productCode = p.productCode
             GROUP BY p.productCode
             ORDER BY product_performance DESC
             LIMIT 10) AS p
)
SELECT productCode, productName, productLine
  FROM products
 WHERE productCode IN low_stock_table AND productCode IN product_performane_table
 LIMIT 10;
```

```python
import matplotlib.pyplot as plt

#matplotlib inline is for using matplotlib visualization inside Jupyter notebook
%matplotlib inline

# Execute a SELECT query
query = ("""
            WITH low_stack_table AS (
                    SELECT p.productCode
                      FROM (SELECT p.productCode, p.productName, od.quantityOrdered, p.quantityInStock,
                                   ROUND(SUM(od.quantityOrdered) * 1.0 / p.quantityInStock, 2) AS low_stock
                              FROM products AS p
                              JOIN orderdetails AS od
                                ON p.productCode = od.productCode
                             GROUP BY p.productCode
                             ORDER BY low_stock DESC) AS p
                                         ),
                product_performance_table AS (
                    SELECT p.productCode
                      FROM (SELECT p.productCode, p.productName, od.quantityOrdered, od.priceEach, p.productLine,
                                   ROUND(SUM(od.quantityOrdered * od.priceEach), 2) AS product_performance
                              FROM orderdetails AS od
                              JOIN products AS p
                                ON od.productCode = p.productCode
                             WHERE p.productCode in (SELECT productCode
                                                       FROM low_stack_table)
                             GROUP BY p.productCode
                             ORDER BY product_performance DESC
                             LIMIT 10) AS p
                                                ),
                priority_products_table AS(
                    SELECT p.productCode, p.productName, p.productLine
                      FROM products AS p
                     WHERE p.productCode IN low_stack_table AND p.productCode in product_performance_table
                                           ),
                productline_count_table AS (
                    SELECT COUNT(productLine) AS productline_count
                      FROM priority_products_table
                                           )
                SELECT productLine, 
                       COUNT(productLine) * 100 / (SELECT productline_count 
                                                     FROM productline_count_table) as percentage
                  FROM priority_products_table
                 GROUP BY productLine
                 ORDER BY productLine DESC
            """)

# Retrieve the results of the query as a pandas dataframe
data = pd.read_sql_query(query, conn)

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(data['percentage'], explode = (0, 0, 0, 0.1), labels=data['productLine'], autopct='%1.0f%%', shadow=True, startangle=90)
ax.legend(title="Product Line",
          loc="center left",
          bbox_to_anchor=(1.2, 0, 0.5, 1))
ax.set_title("Percentage priority products for restocking")

plt.show()
```

According to the pie chart, classic cars are in high demand and are the priority for restocking, with motorcycles closely following. The demand for vintage cars and planes, however, seems to be relatively steady and evenly spread.

### Question 2: How Should We Match Marketing and Communication Strategies to Customer Behavior?

This question involves categorizing customers: finding the VIP (very important person) customers and those who are less engaged.

- VIP customers bring in the most profit for the store.
- Less-engaged customers bring in less profit.

We will start with computing how much profit each customer generates.

```python
Image(url= "https://dq-content.s3.amazonaws.com/600/products-orders-orderdetails-tables.png")
```

```python
%%sql
SELECT o.customerNumber, p.productName, p.productLine,
       ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)), 2) AS profit
  FROM orderdetails AS od
  JOIN products AS p
    ON od.productCode = p.productCode
  JOIN orders AS o
    ON od.orderNumber = o.orderNumber
 GROUP BY o.customerNumber
 ORDER BY profit DESC;
```

#### Finding the VIP and Less Engaged Customers

Finding the top five VIP customers.

```python
%%sql
WITH performance AS (
    SELECT o.customerNumber, p.productName, p.productLine,
           ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)), 2) AS profit
      FROM orderdetails AS od
      JOIN products AS p
        ON od.productCode = p.productCode
      JOIN orders AS o
        ON od.orderNumber = o.orderNumber
     GROUP BY o.customerNumber
     ORDER BY profit DESC
                    )
SELECT c.contactFirstName || ' ' || c.contactLastName AS customer_name,
       c.city, c.country, p.profit
  FROM customers AS c
  JOIN performance AS p
    ON c.customerNumber = p.customerNumber
 LIMIT 5; 
```

Finding the top five least-engaged customers.

```python
%%sql
WITH performance AS (
    SELECT o.customerNumber, p.productName, p.productLine,
           ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)), 2) AS profit
      FROM orderdetails AS od
      JOIN products AS p
        ON od.productCode = p.productCode
      JOIN orders AS o
        ON od.orderNumber = o.orderNumber
     GROUP BY o.customerNumber
     ORDER BY profit
                    )
SELECT c.contactFirstName || ' ' || c.contactLastName AS customer_name,
       c.city, c.country, p.profit
  FROM customers AS c
  JOIN performance AS p
    ON c.customerNumber = p.customerNumber
 LIMIT 5; 
```

We can implement a two-pronged strategy to meet the needs of both customer groups. For VIP customers, we can organise exclusive events and initiatives to foster loyalty and further enhance their satisfaction. For less engaged customers, we can launch special campaigns and initiatives to re-engage them, increasing their interest and commitment to our brand. By tailoring our efforts based on customer categorisation, we can effectively increase loyalty among our VIP customers and reinvigorate engagement with those who are less engaged.

### Question 3: How Much Can We Spend on Acquiring New Customers?

Finding the number of new customers arriving each month. That way we can check if it's worth spending money on acquiring new customers. 

```python
%%sql
WITH payment_with_year_month_table AS (
    SELECT *, 
           CAST(SUBSTR(paymentDate, 1, 4) AS INTEGER) * 100 + CAST(SUBSTR(paymentDate, 6, 7) AS INTEGER) AS year_month
      FROM payments AS p
                                      ),
customers_by_month_table AS (
    SELECT p1.year_month, 
           COUNT(*) AS number_of_customers, 
           SUM(p1.amount) AS total
      FROM payment_with_year_month_table AS p1
     GROUP BY p1.year_month
                            ),
new_customers_by_month_table AS (
    SELECT p1.year_month, 
           COUNT(*) AS number_of_new_customers,
           SUM(p1.amount) AS new_customer_total,
           (SELECT number_of_customers
              FROM customers_by_month_table AS c
             WHERE c.year_month = p1.year_month) AS number_of_customers,
           (SELECT total
              FROM customers_by_month_table AS c
             WHERE c.year_month = p1.year_month) AS total
      FROM payment_with_year_month_table AS p1
     WHERE p1.customerNumber NOT IN (SELECT customerNumber
                                       FROM payment_with_year_month_table AS p2
                                      WHERE p2.year_month < p1.year_month)
     GROUP BY p1.year_month
                                )
SELECT year_month, 
       ROUND(number_of_new_customers*100/number_of_customers,1) AS number_of_new_customers_props,
       ROUND(new_customer_total*100/total,1) AS new_customers_total_props
  FROM new_customers_by_month_table;
```

As we can see, the number of clients has been decreasing since 2003, and in 2004, we had the lowest values. The year 2005, which is present in the database as well, isn't present in the table above, this means that the store has not had any new customers since September of 2004. This means it makes sense to spend money acquiring new customers.

To determine how much money we can spend acquiring new customers, we can compute the Customer Lifetime Value (LTV), which represents the average amount of money a customer generates. We can then determine how much we can spend on marketing.

```python
%%sql
WITH performance AS (
    SELECT o.customerNumber, p.productName, p.productLine,
           ROUND(SUM(od.quantityOrdered * (od.priceEach - p.buyPrice)), 2) AS profit
      FROM orderdetails AS od
      JOIN products AS p
        ON od.productCode = p.productCode
      JOIN orders AS o
        ON od.orderNumber = o.orderNumber
     GROUP BY o.customerNumber
     ORDER BY profit
                    )
SELECT ROUND(AVG(profit), 2) AS LTV
  FROM performance;
```

The Lifetime Value (LTV) indicator provides an insight into the profitability generated by the average customer during their entire relationship with our shop. It allows us to predict future profits. For example, if we acquire ten new customers next month, we can predict that we will earn 390,395 USD. This forecast allows us to make informed decisions about the amount of money we can spend on acquiring new customers.
