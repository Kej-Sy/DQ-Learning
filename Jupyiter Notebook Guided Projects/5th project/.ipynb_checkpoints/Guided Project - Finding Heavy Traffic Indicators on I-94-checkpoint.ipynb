### Traffic Volume: Day vs. Night

We'll start by dividing the dataset into two parts:
- Daytime data: hours from 7 a.m. to 7 p.m. (12 hours),
- Nighttime data: hours from 7 p.m. to 7 a.m. (12 hours).

This is not a perfect criterion for distinguishing between nighttime and daytime, but it's a good starting point.

First we convert `date_time` column to `datetime` type of data using `pd.to_datetime()` function

```python
traffic_I94["date_time"] = pd.to_datetime(traffic_I94["date_time"])
```

Now we use `Series.dt.hour` property to get the hour of every instance of the `date_time` column and:
- isolate the daytime data
- isolate the nighttime data

```python
daytime_data = traffic_I94.copy()[(traffic_I94["date_time"].dt.hour >= 7) & (traffic_I94["date_time"].dt.hour < 19)]
daytime_data.shape
daytime_data
```

```python
nighttime_data = traffic_I94.copy()[(traffic_I94["date_time"].dt.hour >= 19) | (traffic_I94["date_time"].dt.hour < 7)]
nighttime_data.shape
nighttime_data
```

Now we're going to compare the traffic volume at night and during day.

We are going to plot the histograms for both day and night data and look up to few statistic for both datasets using `Series.describe()` function. 

```python
plt.figure(figsize=(11, 4))

plt.subplot(1, 2, 1)
plt.hist(daytime_data["traffic_volume"])
plt.title("Traffic Volume: Day")
plt.xlabel("Traffic Volume")
plt.ylabel("Frequency")
plt.xlim(0, 7500)
plt.ylim(0, 8000)

plt.subplot(1, 2, 2)
plt.hist(nighttime_data["traffic_volume"])
plt.title("Traffic Volume: Night")
plt.xlabel("Traffic Volume")
plt.ylabel("Frequency")
plt.xlim(0, 7500)
plt.ylim(0, 8000)
    
plt.show()
```

```python
daytime_data["traffic_volume"].describe()
```

```python
nighttime_data["traffic_volume"].describe()
```

The histogram that shows the distribution of traffic volume during the day is left skewed. This means that most of the traffic volume values are high. From statistics we see that there are 4,252 or more cars passing the station each hour 75% of the time.

The histogram displaying the nighttime data is right skewed. This means that most of the traffic volume values are low. From statistics we know that 75% of the time, the number of cars that passed the station each hour was less than 2,819. 

Although there are still measurements of over 5,000 cars per hour, the traffic at night is generally light. Our goal is to find indicators of heavy traffic, so we'll only focus on the daytime data moving forward.

### Time Indicators

One of the possible indicators of heavy traffic is time.

We're going to look at a few line plots showing how the traffic volume changed according to the following parameters:
- Month
- Day of the week
- Time of day

We will start checking possible indicators from the month. The fastest way to get the average traffic volume for each month is by using the `DataFrame.groupby()` method.

```python
daytime_data["month"] = daytime_data["date_time"].dt.month
by_month = daytime_data.groupby("month").mean()
by_month["traffic_volume"]
```

```python
by_month["traffic_volume"].plot.line()
plt.xlabel("Months")
plt.ylabel("Average Traffic Volume")
plt.title("Traffic volume changed each month on average")
plt.show()
```

The traffic looks less heavy during cold months (Novemberâ€“February) and more intense during warm months (Marchâ€“October), with one interesting exception: July.

Let's take a look on the traffic volume changed each year in July to chek if traffic is significantly less heavy in that month. 

```python
daytime_data["year"] = daytime_data["date_time"].dt.year
only_july = daytime_data[daytime_data["month"] == 7]
only_july.groupby("year").mean()["traffic_volume"].plot.line()
plt.show()
```

As we can see typically, the traffic is pretty heavy in July. The only exception we see is 2016, which had a high decrease in traffic volume. this may be caused by the reconstruction of the road - [these article from 2016](https://eu.detroitnews.com/story/news/local/detroit-city/2015/06/01/shut-weekend-detroit/28300383/) support this hypothesis.

As a initial conclusion here, we can say that warm months generally show heavier traffic compared to cold months.

 Now we will build line plots for another time unit: day of the week.

```python
daytime_data["day_of_week"] = daytime_data["date_time"].dt.dayofweek
by_dayofweek = daytime_data.groupby("day_of_week").mean()
by_dayofweek["traffic_volume"] #0 is monday, 6 is Sunday
```

```python
by_dayofweek["traffic_volume"].plot.line()
plt.xlabel("Days")
plt.ylabel("Average Traffic Volume")
plt.title("Traffic volume changed each day of week on average")
plt.show()
```

Traffic volume is significantly heavier on business days (Monday â€“ Friday) and lighter on weekends (with values below 4,000 cars).

We'll now generate a line plot for the time of day. The weekends, however, will drag down the average values, so we're going to look at the averages separately. To do that, we'll start by splitting the data based on the day type: business day or weekend.

```python
daytime_data['hour'] = daytime_data['date_time'].dt.hour
bussiness_days = daytime_data.copy()[daytime_data['day_of_week'] <= 4]
weekend = daytime_data.copy()[daytime_data['day_of_week'] >= 5] # 5 == Saturday
by_hour_business = bussiness_days.groupby('hour').mean()
by_hour_weekend = weekend.groupby('hour').mean()

print(by_hour_business['traffic_volume'])
```

```python
print(by_hour_weekend['traffic_volume'])
```

```python
plt.figure(figsize=(12, 3))

plt.subplot(1, 2, 1)
by_hour_business['traffic_volume'].plot.line()
plt.title("Traffic Volume By Hour: Bussines Day")
plt.xlabel("Hours")
plt.ylabel("Traffic Volume")
plt.xlim(6,20)
plt.ylim(1500,6500)

plt.subplot(1, 2, 2)
by_hour_weekend['traffic_volume'].plot.line()
plt.title("Traffic Volume By Hour: Weekend")
plt.xlabel("Hours")
plt.ylabel("Traffic Volume")
plt.xlim(6,20)
plt.ylim(1500,6500)

plt.show()
```

At each hour of the day, the traffic volume is generally higher during business days compared to the weekends. As somehow expected, the rush hours are around 7 and 16 â€” when most people travel from home to work and back.

To summarize, we found a few time-related indicators of heavy traffic:

- The traffic is usually heavier during warm months (Marchâ€“October) compared to cold months (Novemberâ€“February).
- The traffic is usually heavier on business days compared to weekends.
- On business days, the rush hours are around 7 and 16.


### Weather Indicators

Another possible indicator of heavy traffic is weather. The dataset provides us with a few useful columns about weather: `temp`, `rain_1h`, `snow_1h`, `clouds_all`, `weather_main`, `weather_description`.

We starst with colums which are numerical and find their corralation values with `traffic_volume`.

```python
daytime_data.corr()["traffic_volume"]
```

Temperature shows the strongest correlation with a value of just +0.13. The other relevant columns (rain_1h, snow_1h, clouds_all) don't show any strong correlation with traffic_value.

Let's generate a scatter plot to visualize the correlation between temp and traffic_volume.

```python
plt.scatter(daytime_data["traffic_volume"], daytime_data["temp"])
plt.ylim(240,320)
plt.show()
```

We can conclude that temperature doesn't look like a solid indicator of heavy traffic.

Let's now look at the other weather-related columns: `weather_main` and `weather_description`.

### Weather Types

In previous examination of correlation between `trafic_volume` and numerical weather column we didn't find any reliable indicator of heavy treaffic. Now we will look at categorical weather-related columns to check if we can find more useful data.

We will calculate the average traffic volume associaded with each unique value in colums: `weather_main` and `weather_description`.

```python
by_weather_main = daytime_data.groupby('weather_main').mean()
by_weather_description = daytime_data.groupby('weather_description').mean()
```

From dataset `by_weather_main` we take `traffic_volume` colum and it is used to plot figure of bar horizontal type.

Next we add label on x-axis with string "Mean Traffic Volume",

Finaly we show the plot.

```python
by_weather_main["traffic_volume"].plot.barh()
plt.xlabel("Mean Trafic Volume")
plt.show()
```

As we can see on aboue horizontal bar plot there's no weather type where traffic volume exceeds 5,000 cars. This makes finding a heavy traffic indicator more difficult.

From dataset `by_weather_description` we take `traffic_volume` colum and it is used to plot figure of bar horizontal type. We also add a parameter `figsize` to make a plot more readable.

Next we add label on x-axis with string "Mean Traffic Volume",

Finaly we show the plot.

```python
by_weather_description["traffic_volume"].plot.barh(figsize=(6, 10))
plt.xlabel("Mean Traffic Volume")
plt.show()
```

In abowe horizontal bar plot we can see that there are three weather types where trafic volume is more than 5,000 coras:
- light rain and snow,
- shower snow. and
- proximity thunderstorm with drizzle.

These is not clera why  these weather types have the highest average traffic, because these are not so bad weather types. It can be coused by giving up walking or riding a bike in bad weather in favour of driving.
