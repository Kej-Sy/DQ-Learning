## Most Used and Most Viewed

Now we will focus on determining the most popular tags. We're going to create two proxies:
- most used - for each tag we'll count how many time the tag was used 
- most viewed - for each tag we'll count how many time a question with it was viewed.

To do that we'll:
- caculate total count for each tag and transform tag_count to a data frame
- count how many times the tag was used, and
- count how many times a question with that tag was viewed
- Create visualizations for the top tags of each proxy

````python
# Creating a dictionary with information about each tag frequency
tag_count = dict()

for tags in questions['Tags']:
    
    for tag in tags:
        
        if tag in tag_count:
            tag_count[tag] += 1
        else:
            tag_count[tag] = 1
````

````python
# Converting tag_count dictionary to DataFrame
tag_count = pd.DataFrame.from_dict(tag_count, orient='index')
tag_count.head()
````

````python
# Rename column in tag_count dataframe
tag_count.rename(columns={0: "Count"}, inplace=True)
tag_count.head()
````

Let's now sort this dataframe by Count and visualize the top 20 results.

````python
# tag_count DataFrame sorting to find most used
most_used = tag_count.sort_values(by="Count").tail(20)
most_used
````

We set our threshold to 20 tags. As we can see in abowe tabe popularity of the tags declines. This mean that looking at these tags should be enough to help us with our goal. Let's visualize these data.

````python
# Visualization of most used tags
most_used.plot(kind="barh", figsize=(16,8))
plt.title('How many time the tag was used ')
plt.xlabel('Count of use')
plt.ylabel('Tags')
````

Some tags are broad and are unlikely to be useful, for example: python, dataset, r. Before we investigate the tags deeper, let's repeat the same process for views.

````python
# Creating a dictionary with information about each tag view frequency
tag_view_count = dict()

for i, row in questions.iterrows():
    
    for tag in row['Tags']:
        
        if tag in tag_view_count:
            tag_view_count[tag] += row['ViewCount']
        else:
            tag_view_count[tag] = row['ViewCount']
````

````python
# Converting tag_view_count dictionary to DataFrame
tag_view_count = pd.DataFrame.from_dict(tag_view_count, orient='index')
tag_view_count.head()
````

````python
# Rename column in tag_count dataframe
tag_view_count.rename(columns={0: "Viewed Count"}, inplace=True)
tag_view_count.head()
````

````python
# tag_view_count DataFrame sorting to find most viewed
most_viewed = tag_view_count.sort_values(by="Viewed Count").tail(20)
most_viewed
````

````python
# Visualization of most viewed tags
most_viewed.plot(kind="barh", figsize=(16,8), color= "red")
plt.title('How many times a question with that tag was viewed')
plt.xlabel('Count of viewed')
plt.ylabel('Tags')
````

Now let's compare these two plot side by side

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches((24, 10))
most_used.plot(kind="barh", ax=axes[0], subplots=True)
most_viewed.plot(kind="barh", ax=axes[1], subplots=True, color= "red")
````

One common way to see how a pair of objects are related is to measure the frequency the two appears together. We will use this as a criterion to evaluate the linkage of a given pair of tags in our study. First let us make a list of all tags. We then create a dataframe where each row and each column will represent a tag to form a n by n matrix.

````python
# Creat a list of all tags
all_tags = list(tag_count.index)

# Find associations for each pair of tags
associations = pd.DataFrame(index=all_tags, columns=all_tags)
associations.iloc[0:4,0:4]
````

We will  fill the dataframe we created with zeroes. Next for each lists of tags in questions["Tags"], we will increment the intervening tags by one. The end result will be a dataframe that for each pair of tags, it tells us how many times they were used together.

````python
# Fill the datframe with zeros
associations.fillna(0, inplace=True)

# Incrementing tag by one
for tags in questions["Tags"]:
    associations.loc[tags, tags] += 1
    
associations.iloc[0:4,0:4]
````

Now we will apply this dataframe to investigate the association of most used tags. The table is huge (20 by 20), so we we'll add some colors.

````python
# Investigate the associations only for most_used tags
relations_most_used = associations.loc[most_used.index, most_used.index]

def style_cells(x):
    helper_df = pd.DataFrame('', index=x.index, columns=x.columns)
    helper_df.loc["time-series", "r"] = "background-color: yellow"
    helper_df.loc["r", "time-series"] = "background-color: yellow"
    for k in range(helper_df.shape[0]):
        helper_df.iloc[k,k] = "color: blue"
    
    return helper_df

relations_most_used.style.apply(style_cells, axis=None)
````

The cells highlighted in yellow tell us that `time-series` was used together with `r` 22 times. The values in blue tell us how many times each of the tags was used. We saw earlier that `machine-learning` was used 2693 times and we confirm it in this dataframe.

Now we use the table to create a heatmap, so it will be much easier to understand. Since the self-self pairs are 100% matched to themselves, we will remove their value first, thus it won't distract our attention in the generated map.

````python
# Remove the diagnal self pairs
for i in range(relations_most_used.shape[0]):
    relations_most_used.iloc[i,i] = pd.np.NaN
````

````python
# Creating heat map
plt.figure(figsize=(12,8))
sns.heatmap(relations_most_used, cmap="Purples", annot=False)
````

As we can observed on abowe heat map the most used tags have the strongest relationships, as given by the dark concentration in the bottom right corner. This may be due to the fact that each of these tags is used frequently, so they are often used together without a strong relationship between them.

### Explore the Ways to Find the Relationships Between Multiple Tags

Another limitation of our analysis is that we only analyze relationships between pairs of tags, not between multiple groups of tags. Many pairs may correlate with each other in complex ways that we cannot detect in our analysis. For example, it may be that a dataset and scikit-learn, analyzed together, have a "strong" relationship with pandas, but each by itself does not.

There are many data mining tools available that show the probability of relationships between data elements in large data sets. Association rule mining is widely used to discover correlations and recognize patterns and has many practical applications. An association rule consists of two parts: predecessor (if) and successor (then). The predecessor is the element found in the data. The consequent is the element found in conjunction with the predecessor. It forms the basis of some machine learning models for analyzing data for patterns or co-occurrences. Learn more about these topic by clicking [here](https://en.wikipedia.org/wiki/Association_rule_learning).

## Just a Fad?

Deep learning is a class of machine learning algorithms that uses multiple layers to progressively extract higher-level features from the raw input. The adjective "deep" in deep learning refers to the use of multiple layers in the network.

In the final stage of this project, we aim to analyze our data and find evidence to support or refute whether interest in deep learning is just a fad or whether progress will continue.

In the file `all_questions.csv` we stored all of the questions ever asked on DSSE, their dates and tags.



In this we will track the interest in deep learning across time. We will:
- Count how many deep learning questions are asked per time period.
- The total amount of questions per time period.
- How many deep learning questions there are relative to the total amount of questions per time period.

````python
# Reading in dataset
all_questions = pd.read_csv("all_questions.csv", parse_dates=["CreationDate"])
````

````python
all_questions.head()
````

````python
all_questions.info()
````

````python
# Cleaning the `Tags` column
all_questions["Tags"] = all_questions["Tags"].str.replace("^<|>$", "").str.split("><")
````

Before we classified deep learning questions, we should decide what tags are conected with deep learning.

The definition of what constitutes a deep learning tag we'll use is: a tag that belongs to the list ["lstm", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"].

After analyzing all the most_used tags, we decided that the ones on below list had the strongest connection to deep learning: 
["lstm", "cnn", "scikit-learn", "tensorflow", "keras", "neural-network", "deep-learning"].

We'll now create a function that assigns 1 to deep learning questions and 0 otherwise; and we use it.

````python
def class_deep_learning(tags):
    
    for tag in tags:
        
        if tag in ["lstm", "cnn", "scikit-learn", "tensorflow", 
                   "keras", "neural-network", "deep-learning"]:
            return 1
    
    return 0
````

````python
# Applying function to all_questions dataset
all_questions['Deep Learning'] = all_questions['Tags'].apply(class_deep_learning)
all_questions.head()
````

The data-science-techonology landscape isn't very dynamic. So we will track it quarterly.

We don't have all the data for the first quarter of 2020, so we'll get rid of those dates.

````python
all_questions = all_questions[all_questions["CreationDate"].dt.year < 2020]
````

Now we will create a column to identify the quarter in which the question was asked.

````python
# Function to identified a quarter
def fetch_quarter(datetime):
    year = str(datetime.year)[-2:]
    quarter = str(((datetime.month - 1) // 3) + 1)
    return "{y}Q{q}".format(y=year, q=quarter)

# Applying function to all_questions
all_questions["Quarter"] = all_questions["CreationDate"].apply(fetch_quarter)
all_questions
````

Now we'll group by quarter and:
- Count the number of deep learning questions.
- Count the total number of questions.
- Compute the ratio between the two numbers above.

````python
# Grouping by quarter and countig the number of deep learning questions and all questions
quarterly = all_questions.groupby('Quarter').agg({'Deep Learning': ['sum', 'size']})
# Coulmns renaming
quarterly.columns = ['Deep Learning Quesetions', 'Total Questions']
````

````python
quarterly
````

````python
# Compute the ratio between the two numbers
quarterly["Deep Learning Rate"] = quarterly["Deep Learning Quesetions"]\
                                /quarterly["Total Questions"]
````

````python
# The following is done to help with visualizations later.
quarterly.reset_index(inplace=True)
quarterly.sample(5)
````

````python
# Data visualization
ax1 = quarterly.plot(x="Quarter", y="Deep Learning Rate",
                    kind="line", linestyle="-", marker="o", color="orange",
                    figsize=(24,12)
                    )

ax2 = quarterly.plot(x="Quarter", y="Total Questions",
                     kind="bar", ax=ax1, secondary_y=True, alpha=0.7, rot=45,  color= "blue")

for idx, t in quarterly["Total Questions"].iteritems():
    ax2.text(idx, t, str(t), ha="center", va="bottom")
xlims = ax1.get_xlim()

ax1.get_legend().remove()

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles=handles1 + handles2,
           labels=labels1 + labels2,
           loc="upper left", prop={"size": 12})


for ax in (ax1, ax2):
    for where in ("top", "right"):
        ax.spines[where].set_visible(False)
        ax.tick_params(right=False, labelright=False)
````
